Concatenamos el medio con el nombre del journey

```python
journeys['medio_journey_name']=journeys['medio'].str.upper()+'_'+journeys['journeyname']
```

Parte uno: transponer el medio_journey_name en columna

```python
journeys_name_medio=journeys[['subscriber_key','open_date','medio_journey_name','unique_click']]
```

```python
journeys_name_medio.head()
```

```python
journeys_name_medio=pd.pivot_table(journeys_name_medio,index=['subscriber_key','open_date'],columns='medio_journey_name',values='unique_click',aggfunc='max')
```

```python
journeys_name_medio=journeys_name_medio.reset_index().fillna(0)
```

```python
journeys_name_medio=journeys_name_medio[journeys_name_medio['open_date']!='nan']
```

```python
journeys_name_medio=journeys_name_medio.rename(columns={'subscriber_key':'student_id','open_date':'day_info'})
```

```python
journeys_name_medio
```

Parte dos: transponer el journeyname en columnas

```python
journeys_name=journeys[['subscriber_key','open_date','journeyname','unique_click']]
```

```python
journeys_name.head()
```

```python
journeys_name=pd.pivot_table(journeys_name,index=['subscriber_key','open_date'],columns='journeyname',values='unique_click',aggfunc='max')
```

```python
journeys_name=journeys_name.reset_index().fillna(0)
```

```python
journeys_name=journeys_name[journeys_name['open_date']!='nan']
```
