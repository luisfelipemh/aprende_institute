```python
import pandas as pd
```

```python
data0 = pd.read_excel("C:/Users/felip/Desktop/Trabajo/video3.xlsx")
```

```python
data1 = data0[["User Id","Start Time (in seconds)","Stop Time (in seconds)"]]
```

```python
data2 = data1.rename(columns={"User Id":"id","Start Time (in seconds)":"start",
                           "Stop Time (in seconds)":"stop"})
```

```python
data2
```

```python
data = data2[data2["start"]!="-"]
```

```python
data
```

```python
def integer(x):
    x = int(float(x))
    return x
```

```python
data["start"]=data["start"].apply(lambda x: integer(x))
```

```python
data
```

```python
def num_entre(a,b):
    x = list(range(a,b+1))
    return x
```

```python
x=data
data["num_ent"]=data.apply(lambda x: num_entre(x["start"],x["stop"]), axis=1)
```

```python
data
```

```python
seg_x_id=pd.DataFrame(data.groupby("id").apply(lambda x: x.sum())["num_ent"])
```

```python
segundos_id=seg_x_id.reset_index()
```

```python
segundos_id
```

```python
for i in segundos_id["num_ent"]:
    set(i)
```

```python
segundos_id
```

```python
segundos_id['num_ent'] = segundos_id.apply(lambda x: list(set(x['num_ent'])), axis=1)
```

```python
segundos_id1=segundos_id
```

```python
segundos_id1['num_ent']=segundos_id1.apply(lambda x: sorted(list(x['num_ent'])), axis=1)
```

```python
segundos_id1
```

```python
#unir listas
lista_concatenada=segundos_id1.apply(lambda x: x.sum())["num_ent"]
```

```python
lista_concatenada
```

```python
duracion=2.25 #inserte duracion del video
N=int(duracion*60)
df=pd.DataFrame({ 'Segundo' : range(0, N + 1 ,1)})
```

```python
df["Engagement"]=df.apply(lambda x: (lista_concatenada.count(x['Segundo'])/len(segundos_id["id"])), axis=1)
```

```python
df.head(n=50)
```

```python
df.to_excel("video3_a.xlsx",sheet_name='Hoja1',index=False)
```
